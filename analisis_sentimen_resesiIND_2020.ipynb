```
import re
from bs4 import BeautifulSoup
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()
twitter_handle = r'@[A-Za-z0-9_]+'                         # remove twitter handle (@username)
url_handle = r'http[^ ]+'                                  # remove website URLs that start with 'https?://'
combined_handle = r'|'.join((twitter_handle, url_handle))  # join
www_handle = r'www.[^ ]+'                                  # remove website URLs that start with 'www.'
punctuation_handle = r'\W+'
```

```
stopwords = set(pd.read_csv('/content/drive/My Drive//kuliah/TA/stopwordskosong.txt', header=0).stopword)
```

```
def process_text(text):
    soup = BeautifulSoup(text, 'lxml')
    souped = soup.get_text()

    try:
        text = souped.decode("utf-8-sig").replace(u"\ufffd", "?")
    except:
        text = souped

    cleaned_text = re.sub(punctuation_handle, " ",(re.sub(www_handle, '', re.sub(combined_handle, '', text)).lower()))
    cleaned_text = ' '.join([word for word in cleaned_text.split() if word not in stopwords])

    return (" ".join([word for word in tokenizer.tokenize(cleaned_text) if len(word) > 1])).strip()
```

```
# ------ Case Folding --------
#Digunakan untuk transform huruf menjadi huruf lower atau kecil agar seragam
dataabc1['sentiment_text'] = dataabc1['text1'].str.lower()

print('Case Folding Result : \n')
print(dataabc1['sentiment_text'].head(20))
print('\n\n\n')

#Fungsi yang digunakan untuk mengganti atau menghapus karakter yang spesial, hal ini dibutuhkan sebagai bentuk menyamakan pola tulisan agar seragam dan mudah dibaca
def remove_text_special(sentiment_text):
    #hapus tab, new line, and back slice
    sentiment_text = sentiment_text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"").replace('\x80¦',"")
    #hapus non ASCII (emoticon, chinese word, dll)
    sentiment_text = sentiment_text.encode('ascii', 'replace').decode('ascii')
    return sentiment_text

dataabc1['sentiment_text'] = dataabc1['sentiment_text'].apply(remove_text_special)

#hapus emoji ✍ 🌷 📌 👈🏻 🖥
"""
Pada tahapan ini prosesnya dengan meng compile kode emoticon tersebut lalu replace kode tersebut dengan karakter kosong, kode kode tersebut didapatkan dari sumber referensi internet
"""
def remove_emoji(sentiment_text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002500-\U00002BEF"  # chinese char
                           u"\U00002702-\U000027B0"
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u"\U00010000-\U0010ffff"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u200d"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\ufe0f"  # dingbats
                           u"\u3030"
                               #mengonversi data input ke Unicode
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', sentiment_text)

dataabc1['sentiment_text'] = dataabc1['sentiment_text'].apply(remove_emoji)
```

```
def word_tokenize_wrapper(sentiment_text):
    tokens = word_tokenize(sentiment_text)

    #proses normalization menghilangkan Slang Words dari dict appos yang sebelumnya sudah dibuat
    reformed = [slangwords[word] if word in slangwords else word for word in tokens]
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in reformed]
    words = [word for word in stripped if word.isalpha()]
    return words

dataabc1['sentiment_text'] = dataabc1['sentiment_text'].apply(word_tokenize_wrapper)


print('Tokenizing Result : \n')
print(dataabc1['sentiment_text'].head(20))
print('\n\n\n')

"""
Fungsi negation handling digunakan untuk membalikkan polaritas kata kata yang mempengaruhi orientasi sentimen kata lain dalam sebuah kalimat dengan mengubah arti kata tersebut menjadi makna yang sebenarnya. Proses Negation Handling yang dimana setelah di tokenisasi dan koreksi setiap tata Bahasa kata selanjutnya dilakukan proses ini untuk membuat arti dari setiap kata tidak salah makna.
"""
def negation_handling(sentiment_text):
    negation_sentiment_text = []
    for i in range(len(sentiment_text)):
        word = sentiment_text[i]
        if sentiment_text[i-1] not in ['tidak','kurang','tak','tanpa','terlalu','bukan','belum']:
            negation_sentiment_text.append(word)
        else:
            word = "%s_%s" %(sentiment_text[i-1],word)
            negation_sentiment_text.append(word)
    return negation_sentiment_text

dataabc1['sentiment_text'] = dataabc1['sentiment_text'].apply(negation_handling)
print('Negation Result : \n')
print(dataabc1['sentiment_text'].head(20))
print('\n\n\n')
```

```
factory = StemmerFactory() #Variabel factory berisi class untuk proses stemming dari library stemming
stemmer = factory.create_stemmer() #Variabel stemmer digunakan untuk mengambil fungsi pembuatan stemming dari class StemmerFactory()

#fungsi untuk proses stemming
def stemmed_wrapper(term):
  return stemmer.stem(term)

#variabel untuk menampung dictionary dari kata-kata term stemming
term_dict = {}

"""
Proses iterasi untuk pembuatan dict termnya, yang dimana akan dilakukan pengecekan kata kata hasil dari proses tokenisasi, Jadi setelah di tokenisasi, kemudian semua kata-kata tersebut akan ditampung di dalam term_dict yang dimana nantinya kata-kata  tersebut akan melalui proses dari fungsi stemming. Jika kata tersebut bukan termasuk di variabel term_dict maka list kata di  dictionary tersebut akan dikosongkan.
"""
for document in dataabc1['sentiment_text']:
  for term in document:
    if term not in term_dict:
      term_dict[term] = " "

print(len(term_dict))
print("------------------------------")

"""
proses iterasi untuk pengecekan setiap kata di dalam variabel term_dict, yang kemudian di dalam proses tersebut terjadinya proses perubahan setiap kata yang akan dilakukan oleh fungsi stemmed_wrapper yang berisi proses stemming kata. Setelah itu,
akan ditampilkan hasil dari proses perubahan.
"""
for term in term_dict:
  term_dict[term] = stemmed_wrapper(term)
  print(term,":",term_dict[term])

#print(term_dict)
print("-----------------------------")

"""
Kemudian hasil perubahan stemmingnya dimasukkan ke dalam dataframe untuk diolah ke tahapan berikutnya
"""
def get_stemmed_term(document):
  return[term_dict[term] for term in document]

dataabc1['sentiment_text'] = dataabc1['sentiment_text'].apply(get_stemmed_term)
print(dataabc1['sentiment_text'])
```

```
#import stopword
from nltk.corpus import stopwords
stopwords_indonesia = stopwords.words('indonesian')

#import sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

#tokenize
from nltk.tokenize import TweetTokenizer

# Happy Emoticons
emoticons_happy = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
    ])

# Sad Emoticons
emoticons_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
    ])

# all emoticons (happy + sad)
emoticons = emoticons_happy.union(emoticons_sad)

def clean_tweets(tweet):
    # remove stock market tickers like $GE
    tweet = re.sub(r'\$\w*', '', tweet)

    # remove old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)

    # remove hyperlinks
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)

    # remove hashtags
    # only removing the hash # sign from the word
    tweet = re.sub(r'#', '', tweet)

    #remove coma
    tweet = re.sub(r',','',tweet)

    #remove angka
    tweet = re.sub('[0-9]+', '', tweet)

    # tokenize tweets
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_tokens = tokenizer.tokenize(tweet)

    tweets_clean = []
    for word in tweet_tokens:
        if (word not in stopwords_indonesia and # remove stopwords
              word not in emoticons and # remove emoticons
                word not in string.punctuation): # remove punctuation
            #tweets_clean.append(word)
            stem_word = stemmer.stem(word) # stemming word
            tweets_clean.append(stem_word)

    return tweets_clean
```

```
## fungsi untuk melnalankan penilaian atau pembobotan tehadap kata-kata
%%R
score.sentiment = function(kalimat2, kata.positif, kata.negatif, .progress='none')
{
  require(plyr)
  require(stringr)
  scores = laply(kalimat2, function(kalimat, kata.positif, kata.negatif) {
    kalimat = gsub('[[:punct:]]', '', kalimat)
    kalimat = gsub('[[:cntrl:]]', '', kalimat)
    kalimat = gsub('\\d+', '', kalimat)
    kalimat = tolower(kalimat)

    list.kata = str_split(kalimat, '\\s+')
    kata2 = unlist(list.kata)
    positif.matches = match(kata2, kata.positif)
    negatif.matches = match(kata2, kata.negatif)
    positif.matches = !is.na(positif.matches)
    negatif.matches = !is.na(negatif.matches)
    score = sum(positif.matches) - (sum(negatif.matches))
    return(score)
  }, kata.positif, kata.negatif, .progress=.progress )
  scores.df = data.frame(score=scores, clean_text=kalimat2)
  return(scores.df)
}
```
